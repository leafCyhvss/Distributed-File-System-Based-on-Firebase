```python
import pandas as pd
import seaborn as sns
import numpy as np                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline  
```

```python
audi = pd.read_csv("C:/Users/HP/Downloads/Data Analyse/audi.csv")
ford = pd.read_csv("C:/Users/HP/Downloads/Data Analyse/ford.csv")
toyota = pd.read_csv("C:/Users/HP/Downloads/Data Analyse/toyota.csv")
```

```python
print(audi['year'].corr(audi['price']))
 
print(ford['year'].corr(ford['price']))
 

print(toyota['year'].corr(toyota['price']))
```

```python
print(audi.shape)
print(ford.shape)
print(toyota.shape)
```

```python
audi['year'].value_counts()
```

```python
audi["age"] = 2022-audi['year']
ford["age"] = 2022-ford['year']
toyota["age"] = 2022-toyota['year']
```

```python
audi["Current Value"] = audi['price']*(0.875)**audi["age"]
ford["Current Value"] = ford['price']*(0.875)**ford["age"]
toyota["Current Value"] = toyota['price']*(0.875)**toyota["age"]
```

```python
audi = audi.drop(audi[ (audi['year'] <2002 ) | (audi['year'] > 2019) | (audi['Current Value'] <0)].index)
# audi.drop(indexAge , inplace=True)

ford = ford.drop(ford[ (ford['year'] <2002 ) | (ford['year'] > 2019) | (ford['Current Value'] <0)].index)
# ford.drop(indexAge , inplace=True)
# audi.head(15)
toyota = toyota.drop(toyota[ (toyota['year'] <2002 ) | (toyota['year'] > 2019) | (toyota['Current Value'] <0 )].index)
# toyota.drop(indexAge , inplace=True)
```

```python
fig, ax = plt.subplots(1,1)
# sns.countplot(au['Gender'], ax=ax[0])
# sns.countplot(train_data['Dependents'], ax=ax[1])
sns.countplot(y ='year', data = toyota, order=toyota['year'].value_counts()[0:10].index,hue='transmission').tick_params(axis='x', rotation=90)
plt.title("Toyota car production from 2017-2009 based on transmisson wise")
# sns.countplot(y ='year', data = ford, order=ford['year'].value_counts()[0:10].index,hue='transmission').tick_params(axis='x', rotation=90)
fig.savefig('C:/Users/HP/Downloads/Data Analyse/graphs/toyota1.png')
# sns.countplot(x =toyota['year'].value_counts()[0:10], data = toyota, order=toyota['year'],hue='transmission').tick_params(axis='x', rotation=90)
# ax.legend()
```
